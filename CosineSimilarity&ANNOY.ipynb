```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initializing  CountVectorizer with English stop words
vectorizer = TfidfVectorizer()
count_matrix = vectorizer.fit_transform(df['Adapt Company'])
feature_names = vectorizer.get_feature_names_out()

#shape of the count matrix
print("Shape of count matrix:", count_matrix.shape)
print(count_matrix.toarray())

# Display the vocabulary
print(feature_names)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
word_similarity = cosine_similarity(count_matrix, count_matrix)

# Display the similarity matrix
print("Similarity matrix:")
print(word_similarity)
```

```python
df= df.reset_index()
index = pd.Series(df['indices'].values, index=df['R comapny'])
```

```python
def get_recommendations(company_name, cosine_sim=word_similarity):
    # Find the index of the company in the DataFrame
    idx = df[df['R comapny'] == company_name].index[0]

    # Get the pairwise similarity scores of all companies with this company
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the companies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the top 10 most similar companies
    sim_scores = sim_scores[1:11]

    # Get the indices of the similar companies
    company_indices = [i[0] for i in sim_scores]

    # Get the similarity scores of the top 10 most similar companies
    similarity_scores = [i[1] for i in sim_scores]

    # Return the top 10 most similar companies along with their similarity scores
    return list(zip(df['R comapny'].iloc[company_indices], similarity_scores))

# Example usage
company_name = 'uplead.com'
recommendations = get_recommendations(company_name)
for company, score in recommendations:
    print(f"Company: {company}, Similarity Score: {score}")
```

```python
# Initialize Annoy index
annoy_index = AnnoyIndex(count_matrix.shape[1], 'angular')
```

```python
for i in range(55):
    embedding = count_matrix[i].toarray()[0]
    annoy_index.add_item(i, embedding)

# Build the Annoy index
annoy_index.build(10)
```
